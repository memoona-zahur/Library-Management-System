```python
books: list = []
users: list = []

book1: dict = {
    "ID": 1,
    "title": "To Kill a Mockingbird",
    "author": "Harper Lee",
    "genre": "Fiction",
    "status": "Available"
}

book2: dict = {
    "ID": 2,
    "title": "1984",
    "author": "George Orwell",
    "genre": "Dystopian",
    "status": "Available"
}

book3: dict = {
    "ID": 3,
    "title": "The Great Gatsby",
    "author": "F. Scott Fitzgerald",
    "genre": "Classic",
    "status": "Available"
}

book4: dict = {
    "ID": 4,
    "title": "The Catcher in the Rye",
    "author": "J.D. Salinger",
    "genre": "Coming-of-Age",
    "status": "Available"
}

book5: dict = {
    "ID": 5,
    "title": "Moby Dick",
    "author": "Herman Melville",
    "genre": "Adventure",
    "status": "Available"
}


user1: dict = {
    "ID": 1,
    "name": "Alice",
    "borrowed_books": []
}

user2: dict = {
    "ID": 2,
    "name": "Bob",
    "borrowed_books": []
}

user3: dict = {
    "ID": 3,
    "name": "Charlie",
    "borrowed_books": []
}

user4: dict = {
    "ID": 4,
    "name": "Tania",
    "borrowed_books": []
}

user5: dict = {
    "ID": 5,
    "name": "Mishal",
    "borrowed_books": []
}


books.append(book1)
books.append(book2)
books.append(book3)
books.append(book4)
books.append(book5)

users.append(user1)
users.append(user2)
users.append(user3)
users.append(user4)
users.append(user5)
```

```python
def view_all_books():
    print("\nAll Books:")
    for book in books:
        print(f"{book["ID"]}. \"{book["title"]}\" by {book["author"]} ({book["status"]})")
```

```python
def add_book():
    book_id = int(input("Enter book ID: "))
    title = input("Enter book title: ")
    author = input("Enter book author: ")
    genre = input("Enter book genre: ")
    status = "Available"

    new_book: dict = {
        "ID": book_id,
        "title": title,
        "author": author,
        "genre": genre,
        "status": status
    }

    books.append(new_book)
    print(f"\nBook \"{title}\" added successfully.")
```

```python
def view_all_users():
    print("All Users:")
    for user in users:
        print(f"\n{user["ID"]}. {user["name"]} (Books Borrowed: {len(user["borrowed_books"])})")
        if len(user["borrowed_books"]) > 0:
            print("Borrowed Books:")
            for book_id in user["borrowed_books"]:
                borrowed_book = next((book for book in books if book["ID"] == book_id), None)
                if borrowed_book:
                    print(f" - {borrowed_book["title"]}")
        else:
            print("\nNo books borrowed.")
```

```python
def borrow_book():
    user_id = int(input("Enter your User ID: "))
    book_id = int(input("Enter the Book ID you want to borrow: "))
    
    user = next((user for user in users if user["ID"] == user_id), None)
    
    book = next((book for book in books if book["ID"] == book_id), None)

    if user is None:
        print("Invalid User ID.")
        return

    if book is None:
        print("Invalid Book ID.")
        return

    if  book["status"] == "Available":
        book["status"] = "Checked Out"
        user["borrowed_books"].append(book["ID"])
        print(f"\nYou have successfully borrowed \"{book["title"]}\".")
    else:
        print(f"\nSorry, the book \"{book["title"]}\" is currently checked out.")
```

```python
def return_book():
    user_id = int(input("Enter your User ID: "))
    book_id = int(input("Enter the Book ID you want to return: "))

    user = next((user for user in users if user["ID"] == user_id), None)

    if user is None:
        print("Invalid User ID.")
        return

    if book_id not in user["borrowed_books"]:
        print("\nYou do not have this book borrowed.")
        return

    book = next((book for book in books if book['ID'] == book_id), None)
    book["status"] = "Available"
    user["borrowed_books"].remove(book_id)
    print(f"\nYou have successfully returned \"{book["title"]}\".")
```

```python
def search_book():
    search_term = input("Enter the title, author, or genre of the book: ")
    print("Matching Books:")
    found_books = False

    for book in books:
        if (search_term.lower() in book["title"].lower() or
            search_term.lower() in book["author"].lower() or
            search_term.lower() in book["genre"].lower()):
            print(f"\n{book["ID"]}. \"{book["title"]}\" by {book["author"]} ({book["status"]})")
            found_books = True

    if not found_books:
        print("\nNo matching books found.")
```

```python
def view_available_books():
    print("\nAvailable Books:")
    found_books = False
    for book in books:
        if book["status"] == "Available":
            print(f"{book["ID"]}. \"{book["title"]}\" by {book["author"]} ({book["status"]})")
            found_books = True
    if not found_books:
        print("\nNo available books found.")
```

```python
def view_checked_out_books():
    print("\nChecked-Out Books: ")
    found_books = False
    for book in books:
        if book["status"] == "Checked Out":
            print(f"{book["ID"]}. \"{book["title"]}\" by {book["author"]} ({book["status"]})")
            found_books = True
    if not found_books:
        print("\nNo checked-out books found.")
```

```python
def main_menu():
    print("Welcome to the Community Library System!")
    print("----------------------------------------")
    while True:
        try:
           print("\nPlease choose an option:")
           print("1. View all books")
           print("2. Search for a book")
           print("3. View available books")
           print("4. View checked-out books")
           print("5. Borrow a book")
           print("6. Return a book")
           print("7. View all users")
           print("8. Add a new book")
           print("9. Exit")

           choice = input("Enter your choice (1-9): ")
           print(f"Enter your choice (1-9): {choice}")

           if choice == '1':
            view_all_books()
           elif choice == '2':
            search_book()
           elif choice == '3':
            view_available_books()
           elif choice == '4':
            view_checked_out_books()
           elif choice == '5':
            borrow_book()
           elif choice == '6':
            return_book()
           elif choice == '7':
            view_all_users()
           elif choice == '8':
            add_book()
           elif choice == '9':
            print("\nExiting the system. Goodbye!")
            break
           else:
            print("\nInvalid choice. Please try again.")
        except ValueError:
            print("\nPlease enter a valid number.")
```

```python
main_menu()
```
